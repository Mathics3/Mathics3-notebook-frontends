```python
(* A way to randomly draw the Sierpinski Triangle. *)
vertices = {{0,0}, {1,0}, {.5, .5 Sqrt[3]}};
points = NestList[.5(vertices[[ RandomInteger[{1,3}] ]] + #)&, {0.,0.}, 1000];
Graphics[Point[points], ImageSize->Large]
```

```python
(* Another way to draw the same Sierpinski Triangle using in character form using subscripts and superscripts *)
Nest[Subsuperscript[#,#,#]&,0,5]
```

```python
(* An example involving function evaluation. The Collatz function appears to converge to 1, but there is no proof of this. *)
collatz[1] := 1;
collatz[x_ ? EvenQ] := x / 2;
collatz[x_] := 3 x + 1;
list = FixedPointList[collatz, 14];
ListLinePlot[list]
```

```python
lindenmayer[i_, b_, h_, j_, r_, n_] := 
  (a = h; p = j; s = k = {}; t = Flatten;
    (Switch[#,
      6, s = {a, p, s},
      8, {a, p, s} = s,
      _C, k = {k, Line@{p, p += {Cos@a, Sin@a}}}];
     If[# < 9, a += I^# b ]) & /@ t@Nest[# /. r &, i, n];
    Graphics@t@k);
lindenmayer[{C[1], X}, Pi/2, 0, {0, 0}, {X -> {X, 4, Y, C[1]}, Y -> {C[1], X, 2, Y}}, 7]
```

```python
%py 
def fib(n: int) -> int:
    if 0 <= n <= 1:
        return 1
    if n < 0: return -fib(-n)
    return n * fib(n - 1)
```
